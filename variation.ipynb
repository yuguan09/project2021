```python
import matplotlib.pyplot as plt

x = np.arange(-10, 10, 0.1)
 
y =2.72**-(x**2)
 
plt.title("y =e^(-(x^2))")
plt.plot(x, y)
 
plt.show()
```

```python
E=np.zeros(N,dtype='float64')

for i in range(0,N):
    E[i]=random.uniform(-2.5,2.5)
    
int=np.zeros(N,dtype='float64')

sum=0

for i in range(0,N):
    int[i]=2.72**-((alpha*E[i])**2)
    sum+=int[i]
    
#print(sum)


ave=(5*sum)/N
print(ave)
```
